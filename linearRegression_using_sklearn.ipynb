```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/USA_Housing .csv')
df.head()
```

 data shape
 **bold text**

```python
print('train data',df.shape)
```

```python
df.drop
```

```python
#heatmap
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='tab20c_r')
plt.title('missing data')
plt.show()
```

```python
#
df.drop('Address', axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df.shape
```

**Target variable**

```python
x = df.drop('Price', axis=1)
y = df['Price']
```

```python
#shipt in 0 & 1 format
```

```python
from sklearn import  preprocessing
pre_process = preprocessing.StandardScaler().fit(x)
x_transform= pre_process.fit_transform(x)
```

```python
x_transform.shape
x_transform
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train , y_test =train_test_split(x_transform,y, test_size=.20, random_state=101)
```

```python
y_train.shape
```

**Linear Regression**


**Model training**

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

lr =LinearRegression()
lr.fit(x_train ,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred.shape
print(y_pred)
```

```python
sns.scatterplot(x =y_test, y =y_pred , color='blue', label='Actual data')
plt.plot([min(y_test),max(y_pred)],[min(y_test),max(y_pred)], label='ideal line',color='red')
plt.legend()
plt.show()
```

```python
#combine actual & predicted values in side by side
results=np.column_stack((y_test,y_pred))
print(' Actual value     |  predicted value ')
print('-------------------------')
for actual, predicted in results:
    print(f"{ actual :14.2f} |  { predicted:12.2f}")
```

```python
#actual value - predicted value
residual =actual- y_pred.reshape(-1)
print(residual)
```

```python
sns.distplot(residual, kde=True)
```

Model Evaluation


**Linear regression**

```python
#sorce it
from sklearn.metrics import mean_squared_error
print('linear regression')
print('_______________________________')
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print('mean squared error',mse)
print('Root of mean squared error',rmse)
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

dtr= DecisionTreeRegressor()
dtr.fit(x_train,y_train)

#predicting saleprice using test set
y_pred_rf=dtr.predict(x_test)
DTmes= mean_squared_error(y_pred_rf,y_test)
print('DecisionTreeRegressor:',DTmes)



```

**Random forest**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

rfr= RandomForestRegressor()
rfr.fit(x_train,y_train)

#predicting saleprice using test set
y_pred_rf=rfr.predict(x_test)
rfrmes= mean_squared_error(y_pred_rf,y_test)
print('RandomForestRegressor :',rfrmes)


```

**Gradient Boosting Regression**


```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
gbr= GradientBoostingRegressor()
gbr.fit(x_train,y_train)

#predicting saleprice using test set
y_pred_rf=gbr.predict(x_test)
gbrmes= mean_squared_error(y_pred_rf,y_test)
print('GradientBoostingRegressor :',gbrmes)

```

```python
# sample mode scores
model_socores ={
      'linear regression :': 10100187858.867598 ,
     'DecisionTreeRegressor:':32273302906.977055,
     'RandomForestRegressor :':15087503280.464321,
     'GradientBoostingRegressor':12398259808.94909

}

sorted_scores =sorted(model_socores.items() , key=lambda x:x[1])

print('model ranking  to lower to better')
for rank,(model_name, score) in enumerate(sorted_scores, start=1):
  print(f'{rank}, {model_name}: {score}')
```
